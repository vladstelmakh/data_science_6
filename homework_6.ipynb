```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

# Завдання 1 - Завантаження даних

```python
df_data_2d = pd.read_csv('data_2d.csv', header=None)
df_mnist = pd.read_csv('mnist.csv', header=None)
```

# Завдання 1 - K-means та ліктевий метод для data_2d.csv

```python
inertia_values_data_2d = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_data_2d)
    inertia_values_data_2d.append(kmeans.inertia_)
```

# Завдання 1 - Відображення ліктевого методу для data_2d.csv

```python
plt.plot(range(1, 11), inertia_values_data_2d, marker='o')
plt.title('Elbow Method for data_2d.csv')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

# Завдання 1 - Вибір оптимальної кількості кластерів для data_2d.csv

```python
optimal_k_data_2d = 3  # Вибираємо оптимальну кількість кластерів
print(f'Optimal number of clusters for data_2d.csv: {optimal_k_data_2d}')
```

# Завдання 1 - K-means для data_2d.csv з оптимальною кількістю кластерів

```python
kmeans_data_2d = KMeans(n_clusters=optimal_k_data_2d, random_state=42)
labels_data_2d = kmeans_data_2d.fit_predict(df_data_2d)
```

# Завдання 1 - Відображення результатів кластеризації для data_2d.csv

```python
plt.scatter(df_data_2d.iloc[:, 0], df_data_2d.iloc[:, 1], c=labels_data_2d, cmap='viridis')
plt.title('Clustering Result for data_2d.csv')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
inertia_values_mnist = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_mnist)
    inertia_values_mnist.append(kmeans.inertia_)
```

```python
optimal_k_mnist = 4  # Вибираємо оптимальну кількість кластерів
print(f'Optimal number of clusters for mnist.csv: {optimal_k_mnist}')
```

# Завдання 1 - K-means для mnist.csv з оптимальною кількістю кластерів

# Завдання 2 - Візуалізація для data_2d.csv

```python
optimal_k_data_2d = 3  # обрана оптимальна кількість кластерів
kmeans_data_2d = KMeans(n_clusters=optimal_k_data_2d, random_state=42)
labels_data_2d = kmeans_data_2d.fit_predict(df_data_2d)

plt.scatter(df_data_2d.iloc[:, 0], df_data_2d.iloc[:, 1], c=labels_data_2d, cmap='viridis')
plt.title('Clustering Result for data_2d.csv')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

# Завдання 2 - Візуалізація для mnist.csv

```python
optimal_k_mnist = 4  # обрана оптимальна кількість кластерів
kmeans_mnist = KMeans(n_clusters=optimal_k_mnist, random_state=42)
labels_mnist = kmeans_mnist.fit_predict(df_mnist)

# Використання PCA для зменшення розмірності
pca = PCA(n_components=2)
mnist_pca = pca.fit_transform(df_mnist)

plt.scatter(mnist_pca[:, 0], mnist_pca[:, 1], c=labels_mnist, cmap='viridis')
plt.title('Clustering Result for mnist.csv')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
